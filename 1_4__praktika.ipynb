```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = 1

rayleigh_arr = np.random.rayleigh(q, 4000)
theory_arr = np.arange(0, max(rayleigh_arr), 0.0001)
label_r = f"np.random.rayleigh({q}, 200)"
label_t = f"(x/q**2) * np.exp(-x**2/(2*q**2)), q = {q}"

plt.figure(figsize=(10, 5))
plt.title("Плотность распределения Рэлэя")
plt.hist(rayleigh_arr, 50, label=label_r, density=True)
plt.plot(theory_arr, (theory_arr/q**2) * np.exp(-theory_arr**2/(2*q**2)), label=label_t)
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
```

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

with open('dataset.txt', mode='r') as file:
  for human in file.readlines():

    age.append(int(human.split(',')[0]))

    if human.split(',')[1] == 'male':
      sex.append(1)
    else:
      sex.append(0)

    bmi.append(float(human.split(',')[2]))

    children.append(int(human.split(',')[3]))

    if human.split(',')[4] == 'yes':
      smoker.append(1)
    else:
      smoker.append(0)

    if human.split(',')[5] == 'northwest':
      region.append(0)
    elif human.split(',')[5] == 'northeast':
      region.append(1)
    elif human.split(',')[5] == 'southwest':
      region.append(2)
    elif human.split(',')[5] == 'southeast':
      region.append(3)
    else:
      region.append(human.split(',')[5])

    charges.append(float(human.split(',')[6]))

age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
charges = np.array(charges)
```

```python
plt.scatter(age[charges < 15000  ], charges[charges < 15000], label="charges < 15000 - корреляция")
plt.scatter(age[(charges < 15000) & (charges > 10000) & (age < 45)], charges[(charges < 15000) & (charges > 10000) & (age < 45)], label="10000 < charges < 15000, age < 45 - выбросы")

plt.title('Корреляция, Выбросы')
plt.grid()
plt.legend()
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
plt.scatter(bmi[(age > 35) & (age < 50)], charges[(age > 35) & (age < 50)])

plt.scatter(bmi[(age > 35) & (age < 50) & (charges > 18000) & (charges < 30000)], charges[(age > 35) & (age < 50) & (charges > 18000)  & (charges < 30000)], label="35 < age < 50, 18000 < charges < 29000 - кластер")

plt.scatter(bmi[(age > 35) & (age < 50) & (charges < 12000)], charges[(age > 35) & (age < 50) & (charges < 12000)], label="35 < age < 50, charges < 12000 - кластер")

plt.title('Кластеры')
plt.grid()
plt.legend()
plt.xlabel('bmi')
plt.ylabel('charges')
plt.show()
```

```python
plt.scatter(age[(charges < 4000) & (age < 30)], charges[(charges < 4000) & (age < 30)], label="age < 30, charges < 4000 - типичные значения")

plt.title('Типичные значения')
plt.grid()
plt.legend()
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
plt.scatter(bmi[(charges > 30000) & (bmi > 25)], charges[(charges > 30000) & (bmi > 25)], label="age > 25, charges > 30000 - независимость")

plt.title('Независимость')
plt.grid()
plt.legend()
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```
