```python
df = pd.read_csv('/content/ds_salaries.csv', index_col='Unnamed: 0')
```

# переименование столбцов

```python
df.rename(columns = {'work_year':'год ваканции', 
'experience_level':'уровень опыта', 
'employment_type':'вид трудоустройства', 
'job_title':'название работы', 
'salary':'зарплата', 
'salary_currency':'зарплата в валюте', 
'salary_in_usd':'зарплата в долларах', 
'employee_residence':'место жительства работника', 
'remote_ratio':'дистанционная работа',
'company_location':'локация компании',
 'company_size':'размер компании'}, 
inplace = True)
df.head()
```

# Средняя и медианная зарплата по группам

```python
df2=df.groupby(['название работы']).agg(['mean','median','count'])['зарплата в долларах']
df2
```

самые высокооплачиваемые топ 10 специальностей, изходя из средних зарплат

---



```python
df2.sort_values(by = 'mean', ascending=False)[:10]
```

# Средняя и медианная зарплата по регионам

```python
df3=df.groupby(['локация компании']).agg(['mean','median','count'])['зарплата в долларах']
df3
```

самые высокооплачиваемые топ 10 стран, изходя из средних зарплат

---



```python
df3.sort_values(by = 'mean', ascending=False)[:10]
```
